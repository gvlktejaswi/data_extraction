```python

import fitz
import openpyxl
import re
from difflib import get_close_matches


def extract_text_by_page(pdf_path):
    doc = fitz.open(pdf_path)
    return [page.get_text() for page in doc]


def get_keywords_from_excel(excel_path, sheet_name="Sheet1", column_letter="A"):
    wb = openpyxl.load_workbook(excel_path, data_only=True)
    ws = wb[sheet_name]
    keywords = [cell.value.strip() for cell in ws[column_letter] if cell.value]
    wb.close()
    return keywords


def extract_relevant_lines(pages, keywords, fuzziness_threshold=0.6):
    unit_pattern = r"(Pa·s|Pa.s|mPa·s|cP|°C|s⁻¹|Hz|K|%)"
    value_unit_pattern = rf"\d+\.?\d*\s?{unit_pattern}"
    results = {}

    for keyword in keywords:
        keyword_lower = keyword.lower()
        best_line = ""
        best_page = -1
        best_score = 0

        for i, page_text in enumerate(pages):
            if keyword_lower not in page_text.lower():
                continue

            lines = page_text.split("\n")
            for line in lines:
                if "properties addressed" in line.lower() or "instructions" in line.lower():
                    continue

                words = re.findall(r"\w+", line.lower())
                match_score = 0


                if get_close_matches(keyword_lower, words, n=1, cutoff=fuzziness_threshold):
                    match_score += 1

                if re.search(value_unit_pattern, line):
                    match_score += 1.5

                if match_score > best_score:
                    best_score = match_score
                    best_line = line.strip()
                    best_page = i

        if best_line:
            results[keyword] = f"[Page {best_page+1}] {best_line}"
        else:
            results[keyword] = "No relevant value/unit found"

    return results

def main(pdf_path, excel_path):
    pages = extract_text_by_page(pdf_path)
    keywords = get_keywords_from_excel(excel_path)
    kv_results = extract_relevant_lines(pages, keywords)

    print("Extracted Keyword Contexts:")
    for key, value in kv_results.items():
        print(f"{key}: {value}")

pdf_path = "mark-polymer-data-handbook.pdf"
excel_path = "5.6Rheological Properties.xlsx"
main(pdf_path, excel_path)
```

```python
!pip install pdfplumber
import pdfplumber
import pandas as pd
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import json


excel_df = pd.read_excel("5.6Rheological Properties.xlsx")

model = SentenceTransformer("all-MiniLM-L6-v2")

def extract_pdf_chunks(pdf_path, lines_per_chunk=5):
    chunks, page_map = [], []
    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            text = page.extract_text()
            if not text:
                continue
            lines = text.split('\n')
            for j in range(0, len(lines), lines_per_chunk):
                chunk = " ".join(lines[j:j+lines_per_chunk]).strip()
                if chunk:
                    chunks.append(chunk)
                    page_map.append(i + 1)
    return chunks, page_map

def build_faiss_index(chunks):
    embeddings = model.encode(chunks, convert_to_numpy=True)
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings)
    return index, embeddings

def semantic_search_with_filtering(keywords, chunks, page_map, index, top_k=20):
    results = {}
    for keyword in keywords:
        q_embedding = model.encode([keyword])
        D, I = index.search(np.array(q_embedding), top_k)

        matches = []
        for idx in I[0]:
            chunk = chunks[idx]
            if keyword.lower() in chunk.lower():
                page = page_map[idx]
                matches.append(f"Page {page}: {chunk}")

        if not matches and len(I[0]) > 0:
            fallback_chunk = chunks[I[0][0]]
            fallback_page = page_map[I[0][0]]
            matches.append(f"Page {fallback_page}: {fallback_chunk} (fallback match)")

        results[keyword] = matches
    return results

def run_semantic_keyword_extraction(pdf_path, excel_df):
    keywords = excel_df.iloc[:, 0].dropna().astype(str).tolist()
    chunks, page_map = extract_pdf_chunks(pdf_path)
    index, _ = build_faiss_index(chunks)
    results = semantic_search_with_filtering(keywords, chunks, page_map, index)
    print(json.dumps({"5.6": results}, indent=2, ensure_ascii=False))
    return results


results = run_semantic_keyword_extraction("mark-polymer-data-handbook.pdf", excel_df)
```
